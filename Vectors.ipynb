```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
### Algebraic and geometric interpretations of vectors

# 2D vector
v2 = np.array([ 4, 3 ])

# 3D vector
v3 = np.array([ 4, -3, 2 ])

# Row to column
v3t = np.transpose(v3)

## Plotting 2D
# Plot x1,x2   y1,y2
plt.plot([0,v2[0]],[0,v2[1]])
# X and y axis equal
plt.axis('equal')
# Plots a horizontal dashed line at y=0 from x=-4 to x=4
plt.plot([-4, 4],[0, 0],'k--')
# Plots a vertical dashed line at x=0 from y=-4 to y=4
plt.plot([0, 0],[-4, 4],'k--')
plt.grid()
# Limits of the grid
plt.axis((-4, 4, -4, 4))
# Display plot
plt.show()

## Ploting 3D
#  Create a new figure with an aspect ratio of 1
fig = plt.figure(figsize=plt.figaspect(1))
# Type of projection
ax = plt.axes(projection='3d')
# Plot x1,x2  y1,y2  z1,z2
ax.plot([0,v3[0]], [0,v3[1]], [0,v3[2]], linewidth=3)
# Plots a horizontal dashed line at y=0 z=0 from x=-4 to x=4
plt.plot([-4, 4],[0, 0],[0, 0],'k--')
# Plots a vertical dashed line at x=0 z=0 from y=-4 to y=4
plt.plot([0, 0],[-4, 4],[0, 0],'k--')
# Plots a transfersal dashed line at x=0 y=0 from z=-4 to z=4
plt.plot([0, 0],[0, 0],[-4, 4],'k--')
# Display plot
```
